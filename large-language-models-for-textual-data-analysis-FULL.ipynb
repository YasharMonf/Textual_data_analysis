## Introduction

In our previous session, we explored TextBlob, a library that treats language like a set of rules and dictionaries. It was fast and effective for basic tasks, but it often missed the "soul" of the conversation including sarcasm, complex context, and cultural nuances. They were also processing a body of text sequentially, which can be time-consuming when dealing with a large volume of work, such as a book.

Today, we move from Rule-Based NLP to Generative AI. Using Large Language Models (LLMs) with Python allows us to stop treating text as just "data" and start treating it as "ideas."

Traditional language models and NLP libraries like TextBlob or NLTK look at words in isolation or small chunks. LLMs, built on the deep neural networks and Transformer architecture, use something called Attention. In simple words, this allows the model to "pay attention" to the beginning of a book while it’s reading the final chapter, understanding how every word relates to every other word. Attention Mechanism is the foundation that allows AI to understand context. Imagine you are at a loud party. There are dozens of people talking at once (this is like a massive dataset of words). Without Attention, you try to listen to every single sound in the room equally. You quickly become overwhelmed and understand nothing. With Attention, you lock eyes with one friend. Your brain "muffles" the background music and the other conversations. You focus your mental energy on their specific words because they are the most relevant to you right now.

In a sentence, the AI does the same thing. For every word it processes, it "looks" at all the other words and decides which ones are most important for understanding the current one.

Technically, it uses three mathematical vectors for every word:

- Query (Q): What am I looking for?

- Key (K): What do I offer? 

- Value (V): What information do I actually hold?

The model multiplies these together to decide how much "focus" to put on each word. Attention allows a model to connect a pronoun on page 100 of a book to a name mentioned on page 1 (Long-range memory). Because the model can look at all words at once (instead of one-by-one), it is incredibly fast to train on massive amounts of data (Parallel processing).

## 1) Large Language Models

The launch of ChatGPT by OpenAI highlighted the remarkable progress in LLM technology, which was previously accessible only to researchers and organizations with extensive resources and specialized expertise. For example, while models like GPT-3 had been available via API, ChatGPT’s web-based interface allowed the public to experience and utilize advanced AI capabilities without needing technical knowledge or large-scale computing resources.

The quality and variety of training data have been pivotal in driving the recent success of LLMs. As methods for collecting and processing vast, high-quality datasets improve, these models can achieve far more accurate and contextually relevant responses.

Advances in powerful computing resources, particularly graphics processing units (GPUs) and, more recently, tensor processing units (TPUs), alongside improved data processing techniques, have revolutionized the training of LLMs.

A major breakthrough has been the integration of human feedback directly into the training process, driving some of the most substantial gains in model capabilities. For example, models like ChatGPT have been enhanced using Reinforcement Learning from Human Feedback (RLHF), where human evaluators rate model responses. This feedback is then used to fine-tune the model, making its answers more accurate, helpful, and aligned with human expectations.

Based on license, we have different types of LLMs.

**Proprietary Models:**
Examples: GPT-5, Gemini 3.
Characteristics: Limited Access, API-based use (The source code, training strategies, model weights, and even details like the number of parameters they have are all kept secret.)

**Open-source Models:**
Example: BERT
Characteristics: Downloadable, Modifiable, Customizable.

**Open licenses with licensing considerations:**
Example: Llama 3, Gemma 2

Open licenses, like those for Llama 3 or Gemma 2, provide more restrictions than open-source licenses. For example, Llama 3's license allows commercial use up to 700 million monthly users but restricts certain use cases, such as prohibiting large tech companies from using it. Gemma 2’s license blocks certain types of applications, like those that promote illegal activities. These limitations help companies maintain some control over how their models are used, while still allowing broad access for smaller developers.

Based on the architecture, we also have different type of LLMs:

**Encoder-Only Models**: These models utilize only the encoder part of the Transformer architecture. They are primarily designed for tasks like text classification (sentiment analysis), finding the relevant answer within a given context., and identifying and classifying named entities (e.g., people, organizations, locations) in text.  Examples: BERT, ALBERT, RoBERTa

**Decoder-Only Models**: These models utilize only the decoder part of the Transformer architecture. They are primarily designed for tasks like text generation, dialogue generation and chat bots. Examples: GPT, Gemini, Llama

**Encoder-Decoder Models**: These models utilize both the encoder and decoder parts of the Transformer architecture.   
They are well-suited sequence-to-sequence learning like translating text from one language to another or 
 speech-to-text transcription. Example: T5

Note: Decoder-only models leverage their powerful language generation capabilities, 
combined with creative prompt engineering and sometimes fine-tuning, can tackle tasks 
that might traditionally require an encoder-decoder architecture.

Let's start with a simple model for **text completion (generation) via LLMs**:

```python
# Text Generation using an LLM
from transformers import pipeline

# Load the pre-trained GPT-2 model for text generation
text_generator = pipeline("text-generation", model="gpt2")

# Prompt for text generation
prompt = "Once upon a time, there was a curious little cat named Whiskers who"

# Generate a story
story = text_generator(prompt, max_new_tokens=100, do_sample=True, temperature=0.7)

# Print the generated story
print("Generated Story:")
print(story[0]['generated_text'])
```

If you run the code on your CPU, this can take a while. To speed things up, you may want to enable an accelerator on Kaggle and select a GPU resource to run your code. This can speed things up significantly, especially when running a larger model or if you try to fine-tune the model.

In the above example, temperature and do_sample are the creative dials of the model. They control how much the AI sticks to the most likely next word versus taking a risk on a less common one.

**do_sample**: This is a boolean parameter (True or False) that determines the strategy the model uses to pick the next word.

- do_sample=False (Greedy Decoding): The model always picks the word with the absolute highest probability. It is predictable and consistent. If you run the same prompt twice, you’ll usually get the exact same answer.

- do_sample=True (Sampling): The model looks at a pool of likely words and picks one randomly based on their probability weights. This introduces variety and "personality."

**temperature**: Temperature only matters when do_sample=True. It mathematically adjusts the probability distribution of the next possible words.

In technical terms, the model calculates raw scores for every word in its vocabulary. Temperature modifies these scores before they are turned into percentages.

***Low Temperature (e.g., 0 – 0.4): "The Focused Professional"***

It makes high-probability words even more likely and crushes low-probability words.

Result: Highly factual, concise, and repetitive. Best for coding or data extraction.

***Medium Temperature (e.g., 0.5 – 0.7): "The Balanced Storyteller"***

This is the "sweet spot" for most chat applications. It allows for some creative flair without losing the plot.

***High Temperature (e.g., 0.7 and above): "The Wild Artist"***

It levels the playing field, making unlikely words almost as likely as common ones.

Result: Highly creative, but prone to "hallucinations," gibberish, or losing track of the instructions.

```python
## adding code for selecting GPU when it is available
from transformers import pipeline

# Importing PyTorch library which gives you access to the cuda.is_available().
import torch

# Check if GPU is available (optional but good practice)
# checks if the system has an NVIDIA GPU and the necessary drivers (CUDA) installed
# the integer 0 refers to the first available GPU. 
# If you had multiple GPUs, you could use 1, 2, etc.
# -1: This is the specific code for the CPU

device = 0 if torch.cuda.is_available() else -1

# Load the model and explicitly move it to the GPU (device 0)
text_generator = pipeline(
    "text-generation", 
    model="gpt2", 
    device=device  # 0 refers to the first GPU, -1 refers to CPU
)

# Prompt for text generation
prompt = "Once upon a time, there was a curious little cat named Whiskers who"

# Generate a story
story = text_generator(prompt, max_new_tokens=100, do_sample=True, temperature=0.7)

print("Generated Story using", 'GPU' if device == 0 else 'CPU', ":")
print(story[0]['generated_text'])
```

### Zero-Shot Topic Classification Example

Topic Classification is a task where a model looks at a piece of text and automatically assigns it a label or category based on its content. It is the digital version of filing a news article into the Sports, Politics, or Tech section of a newspaper. In the past, you had to train a model specifically on thousands of examples of Sports articles before it could recognize one. 

However, with the Zero-Shot learning capability of LLMs, the model has already learned the "logic" of language. You can simply give it a list of labels it has never seen before, and it will use its general knowledge to decide which one fits the text the best. During training, the model learned that words like "Mars," "Rover," and "NASA" are close to the concept of "Tech" and "Space" in its internal brain.


```python
from transformers import pipeline
import torch

# 1. Load the lightweight DeBERTa model
# This model is faster on Kaggle GPUs
model_id = "MoritzLaurer/DeBERTa-v3-base-mnli-fever-anli"

classifier = pipeline("zero-shot-classification", model=model_id, 
    device=0 if torch.cuda.is_available() else -1, # Use GPU if available
    dtype=torch.float16 # Speeds up inference on Kaggle
)

text = "NASA launched a new Mars rover to search for signs of ancient life."
labels = ["space", "politics", "sports", "technology", "health", "history"]

# 2. Perform classification
# multi_label=True allows the model to give high scores to multiple categories
results = classifier(text, labels, multi_label=False)

print("Text:", text)
print("Results:")

# print the results

for i in range(len(results['labels'])):
    label = results['labels'][i]
    score = results['scores'][i]
    
    if score > 0:
        print(label, "=>", score)
```

Here we want to first, implement a traditional language model via TextBlob to perform a sentiment analysis on the following three sample movie reviews (output can be either positive or negative sentiment). #### Sample texts

- "I left the cinema after only 10 minutes!!"



```python
# If you get error, you probably have issues with version mismatch of libraries
# you may need to downgrade your transformer library, and then restart the kernel
# you can copy the commented code into a new code block and run it before running
# the main code in that case
# Here, we force the older version for compatibility
# !pip install --force-reinstall transformers==4.48.0
# import os
# This can kill the process to force a restart
# os._exit(0)

from textblob import TextBlob
from transformers import pipeline

# 1. Setup the Sample Texts
samples = [
    "I loved the movie. It was fantastic!",
    "The movie is terrible. I would not recommend it.",
    "I left the cinema after only 10 minutes!!"
]

# 2. Load the LLM pipeline
sentiment_model = pipeline("sentiment-analysis", 
                 model="distilbert/distilbert-base-uncased-finetuned-sst-2-english")

# 3. Get LLM results for all samples at once
llm_results = sentiment_model(samples)

print("--- RESULTS ---")

# 4. Use a manual range loop to access items by their position (0, 1, 2, 3)
for i in range(len(samples)):
    # Get the text from the samples list
    text = samples[i]
    
    # Get the sentiment score from TextBlob
    tb_score = TextBlob(text).sentiment.polarity
    
    # Get the label and score from the LLM results list
    llm_label = llm_results[i]['label']
    llm_percent = llm_results[i]['score']
    
    # Print using simple commas
    print("TEXT:", text)
    print("TextBlob Polarity:", tb_score)
    print("LLM Label:", llm_label)
    print("LLM Confidence:", llm_percent)
    print("----------------")
```

**Choosing a Model**

Before calling an LLM API, one of the most important engineering decisions is model selection.
The model you choose directly affects cost, latency, and capability, which in turn determines whether your solution is appropriate for learning, prototyping, or production.

In this notebook, we make this choice explicitly and intentionally.

**Key factors when choosing an LLM**

When selecting a model, always reason along three core axes.

1) Cost: LLM usage is billed based on tokens. This includes input tokens (what you send to the model), output tokens (what the model generates), and in some cases, cached or system tokens.

2) Speed (Latency): Latency matters when you are building interactive systems and you want fast feedback during development.

3) Capability: Models differ in capability across dimensions such as reasoning depth, instruction-following quality, context window size and multimodal support (text, images, etc.)

More capable models are powerful, but they are often more expensive, slower, and unnecessary for many common tasks. Choosing more capability than required usually increases cost without adding value.

**Gemini Model Families (High-Level Overview)**

Google provides multiple Gemini model variants designed for different use cases.

**Gemini Pro**

- Higher reasoning and language capabilities
- Better suited for complex tasks and long contexts
- Higher cost and latency
- Typically used in advanced or production scenarios

**Gemini Flash**

- Optimized for speed and cost efficiency
- Lower latency and lower cost per token
- Strong performance for most common tasks

## 4) Next steps: Prompt-Engineering, Fine-Tuning or using RAG!

Now that you were able to use pre-trained LLMs via Hugging Face Transformer or external APIs, the question is what are the next steps. While a base model like Llama 3, Gemini 3.0 or GPT-5 is brilliant at general reasoning, it doesn't know your company's private files, your research group survey data, today’s news, or your specific way of analzying data.

To fix this, you have three primary paths: using advanced prompt-engineering techniques (teaching the model the art of communication), using Retrieval-Augmented Generation or RAG (giving the model a library to look things up) or Fine-Tuning the model (changing the model’s "brain" through extra training). 

**Prompt engineering** is the process of refining the input to guide the model toward a more accurate, useful, or specialized output without changing a single weight in the model's architecture. **Chain-of-Thought (CoT) Prompting** is one of the most effective method of prompt enginnering. Here instead of asking for an answer directly, you instruct the model to "think step-by-step." This forces the LLM to allocate more "compute" (tokens) to the reasoning process before arriving at a conclusion. Another method of prompt-engineering is **Few-Shot prompting**. You provide the model with 3–5 examples of the desired input-output pair. This is often more effective than a 500-word instruction manual. 

Think of **RAG** as giving the model an open-book exam. Instead of memorizing your data, the model searches for the relevant information in an external database and reads it before answering. 

On the other hand, **Fine-tuning** is like sending the model to specialized graduate school. You show it thousands of examples until it learns a specific style, format, or jargon. 

There is also a fourth option. Some production systems use a **Hybrid strategy** combining both RAG and Fine-Tuning with the prompt-engineering techniques! 

## 5) Ethical and Security Considerations

AI models in general can be extremely useful, but they have intrinsic limitations depending on the data they have been trained on. Similarly, LLMs are incredibly powerful but they have several limitations, including hallucinations, bias, limited reasoning, high training costs, and energy consumption.LLMs are usually pretty good at understanding and generating sentences. However, they are not the best tools for solving complex problems. 

In addition to these limitations, using and deploying AI tools, particularly LLMs, require significant attention to ethics as these models are always based on textual data which can be found on the internet and fed into the model by random users (for example text data from wikipedia, Reddit, twitter (X), Facebook, Instagram and weblogs). Some key areas of concern are data privacy, bias, and fairness. For example, concerns about privacy violations and the potential misuse of personal information includes:

- Avoid data memorization

- Avoid disclosure of confidential information

- Comply with relevant privacy laws (obtaining consent, providing notice about data collection)

- Comply with terms of service (ToS) of websites

- Data security


There are different type of security attacks in LLMs. The most important ones are prompt hacking and adversarial attacks. **Prompt hacking** is referred to as a technique that involves manipulating the input of a model to obtain a desired unintended output. It is used to directly manipulate the LLM's behavior, often bypassing safety measures or revealing sensitive information. Two main techniques includes Prompt Injection and Jailbreaking Attacks. Another type of security attacks in LLMs are **adversarial attacks**. An adversarial attack involves manipulating input data to cause the network to produce incorrect or unintended outputs. The goal of these attacks is to mislead or confuse the LLM, causing it to produce incorrect or nonsensical outputs. Two main types are Data poisoning and Backdoor attack.

Defending against prompt hacking and adversarial attacks is an ongoing research area.  Some strategies include:

- Adversarial Training: This involves training the LLM on a dataset that includes adversarial examples (slightly modified inputs designed to fool the model).  This helps the model learn to recognize and resist these attacks. 

- Input Sanitization:  This involves cleaning and preprocessing the input before it's fed to the LLM.

- Prompt Rewriting:  Automatically rewriting or paraphrasing user prompts to remove potentially harmful instructions or ambiguities.

- Sandboxing and Isolation: Restricting the LLM's access to external systems or sensitive data to limit the potential damage from successful attacks
